```python
import pandas as pd
import numpy as np
from prophet import Prophet
import streamlit as st
import plotly.graph_objs as go
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Load the datasets
consumption_file_path = 'data (25).csv'
service_file_path = 'data (27).csv'
consumption_data = pd.read_csv(consumption_file_path)
service_data = pd.read_csv(service_file_path)

# Mapping dictionary
mapping = {
    'MOH 747A_Male Condoms': 'Male Condoms',
    'MOH 711 Client receiving Male condoms': 'Male Condoms',
    'MOH 747A_Female Condoms': 'Female Condoms',
    'MOH 711 Clients receiving Female Condoms': 'Female Condoms',
    'MOH 747A_Emergency Contraceptive pills': 'EC Pills',
    'MOH 711 Emergency contraceptive pill': 'EC Pills',
    'MOH 747A_Combined Oral contraceptive Pills': 'COCs',
    'MOH 711 Pills Combined oral contraceptive': 'COCs',
    'MOH 747A_Progestin only pills': 'POPs',
    'MOH 711 Pills progestin only': 'POPs',
    'MOH 747A_Cycle Beads': 'Cycle Beads',
    'MOH 711 Rev 2020_Clients given cycle beads': 'Cycle Beads',
    'MOH 747A_DMPA-IM': 'DMPA-IM',
    'MOH 711 Rev 2020_FP Injections DMPA- IM': 'DMPA-IM',
    'MOH 747A_DMPA-SC': 'DMPA-SC',
    'MOH 711 Rev 2020_FP Injections DMPA- SC': 'DMPA-SC',
    'MOH 747A_Implants (1-Rod) – ENG 68mg': 'Implanon',
    'MOH 711 Rev 2020_Implants insertion 1 Rod': 'Implanon',
    'MOH 711 Rev 2020_Implants insertion 2 Rod': '2 Rod',
    'MOH 747A_Hormonal IUCD': 'Hormonal IUCD',
    'MOH 711 Rev 2020_IUCD Insertion Hormonal': 'Hormonal IUCD',
    'MOH 747A_Non-Hormonal IUCD': 'Non-Hormonal IUCD',
    'MOH 711 Rev 2020_IUCD Insertion Non Hormonal': 'Non-Hormonal IUCD',
    'MOH 747A_Implants (2-Rod) - LNG 75mg (3 years)': 'Levoplant',
    'MOH 747A_Implant (2-Rod) – LNG 75mg (5 years)': 'Jadelle'
}

# Apply the mapping to the columns
consumption_data.rename(columns=mapping, inplace=True)
service_data.rename(columns=mapping, inplace=True)

# Add a column to distinguish between consumption and service
consumption_data['Type'] = 'Consumption'
service_data['Type'] = 'Service'

# Combine the datasets
combined_data = pd.concat([consumption_data, service_data], ignore_index=True)

# Fill missing values
combined_data.fillna(0, inplace=True)

# Convert 'periodname' to datetime and aggregate monthly
combined_data['periodname'] = pd.to_datetime(combined_data['periodname'], format='%b-%y', errors='coerce')

# Aggregate on a monthly level
combined_data = combined_data.groupby([pd.Grouper(key='periodname', freq='M'), 'organisationunitname', 'Type']).sum().reset_index()

# Remove columns that are not needed in the aggregation (non-numeric ones)
non_numeric_columns = combined_data.select_dtypes(exclude=[np.number]).columns.difference(['periodname', 'organisationunitname', 'Type'])
combined_data.drop(non_numeric_columns, axis=1, inplace=True)
```

```python
# Function to check if there is enough data for seasonal decomposition
def has_sufficient_data(df, min_periods=24):
    return len(df) >= min_periods

# Function to perform time series decomposition
def decompose_time_series(df, column):
    try:
        result = seasonal_decompose(df.set_index('ds')[column], model='additive', period=12)
        return result
    except ValueError:
        return None

# Function to plot trends
def plot_trends(data, product):
    consumption_data = data[data['Type'] == 'Consumption'][['periodname', product]].rename(columns={'periodname': 'ds', product: 'y'})
    service_data = data[data['Type'] == 'Service'][['periodname', product]].rename(columns={'periodname': 'ds', product: 'y'})

    trend_fig = go.Figure()

    # Check if there is sufficient data for decomposition
    if has_sufficient_data(consumption_data):
        consumption_trend = decompose_time_series(consumption_data, 'y')
        if consumption_trend is not None:
            trend_fig.add_trace(go.Scatter(x=consumption_trend.trend.index, y=consumption_trend.trend, mode='lines', name='Consumption Trend', line=dict(color='red')))
        else:
            trend_fig.add_trace(go.Scatter(x=consumption_data['ds'], y=consumption_data['y'], mode='lines', name='Consumption Data', line=dict(color='red')))
    else:
        st.warning("Insufficient data for seasonal decomposition for consumption data.")
        trend_fig.add_trace(go.Scatter(x=consumption_data['ds'], y=consumption_data['y'], mode='lines', name='Consumption Data', line=dict(color='red')))
    
    # Check if there is sufficient data for decomposition
    if has_sufficient_data(service_data):
        service_trend = decompose_time_series(service_data, 'y')
        if service_trend is not None:
            trend_fig.add_trace(go.Scatter(x=service_trend.trend.index, y=service_trend.trend, mode='lines', name='Service Trend', line=dict(color='blue')))
        else:
            trend_fig.add_trace(go.Scatter(x=service_data['ds'], y=service_data['y'], mode='lines', name='Service Data', line=dict(color='blue')))
    else:
        st.warning("Insufficient data for seasonal decomposition for service data.")
        trend_fig.add_trace(go.Scatter(x=service_data['ds'], y=service_data['y'], mode='lines', name='Service Data', line=dict(color='blue')))
    
    trend_fig.update_layout(title=f'Trend for {product}', xaxis_title='Date', yaxis_title='Quantity')
    st.plotly_chart(trend_fig)
```

```python
# Function to train Prophet models and plot forecasts
def train_and_plot_forecasts(data, product, interactive):
    data_product = data[['periodname', product]].rename(columns={'periodname': 'ds', product: 'y'})

    # Check if there is enough data to fit the model
    if len(data_product) < 2:
        st.warning("Insufficient data available to perform forecasting.")
        return
    
    # Train the model
    model = Prophet()
    model.fit(data_product)
    
    # Make future DataFrame for 12 months
    future = model.make_future_dataframe(periods=12, freq='M')
    forecast = model.predict(future)
    
    # Format the date
    forecast['ds'] = forecast['ds'].dt.strftime('%Y-%m-%d')
    
    # Ensure no negative forecasts
    forecast['yhat'] = forecast['yhat'].clip(lower=0)
    forecast['yhat_lower'] = forecast['yhat_lower'].clip(lower=0)
    forecast['yhat_upper'] = forecast['yhat_upper'].clip(lower=0)
    
    if interactive == 'Yes':
        # Plot the forecast with Plotly for interactivity
        forecast_fig = go.Figure()
        forecast_fig.add_trace(go.Scatter(x=data_product['ds'], y=data_product['y'], mode='lines', name='Actual', line=dict(color='blue')))
        forecast_fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], mode='lines', name='Forecast', line=dict(color='red')))
        forecast_fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_lower'], mode='lines', fill='tonexty', line=dict(color='lightgrey'), showlegend=False, name='Lower Bound'))
        forecast_fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_upper'], mode='lines', fill='tonexty', line=dict(color='lightgrey'), showlegend=False, name='Upper Bound'))
        forecast_fig.update_layout(title=f'Forecast for {product}', xaxis_title='Date', yaxis_title='Quantity', hovermode='x unified')
        st.plotly_chart(forecast_fig)
    else:
        # Plot the forecast with Matplotlib for static plot
        plt.figure(figsize=(10, 6))
        plt.plot(data_product['ds'], data_product['y'], label='Actual', color='blue')
        plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red')
        plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='gray', alpha=0.2)
        plt.title(f'Forecast for {product}')
        plt.xlabel('Date')
        plt.ylabel('Quantity')
        plt.legend(loc='upper left')
        plt.grid(True)
        st.pyplot(plt)
    
    # Display forecast table without the index column and with bold headers
    forecast_display = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12)
    forecast_display.columns = ['Date', 'Forecast', 'Lower Bound', 'Upper Bound']
    st.write(forecast_display.to_html(index=False, classes='forecast-table'), unsafe_allow_html=True)
    
    # Calculate total forecasted quantity for the next year
    total_forecast = forecast_display['Forecast'].sum()
    st.write(f'<p style="font-weight:bold;">Total forecasted quantity for the next year: {total_forecast:.2f}</p>', unsafe_allow_html=True)

       
```

```python
# Streamlit app
def main():
    st.set_page_config(layout="wide", page_title="Prophet Forecasting App")
    st.markdown(
        """
        <style>
        .main {
            background-color: #f0f0f5;
            padding: 20px;
            font-family: Arial, sans-serif;
        }
        .forecast-table th {
            font-weight: bold;
            text-align: center;
        }
        .forecast-table td {
            text-align: center;
        }
        </style>
        """,
        unsafe_allow_html=True
    )
    st.markdown('<div class="main">', unsafe_allow_html=True)
    st.title('Prophet Forecasting App')
    
    st.sidebar.header('Configuration')
    organization_selection = st.sidebar.selectbox('Select Organization', combined_data['organisationunitname'].unique())
    product_selection = st.sidebar.selectbox('Select Product', combined_data.columns[3:].unique())
    data_type_selection = st.sidebar.selectbox('Select Data Type', ['Consumption', 'Service', 'Both'])
    
    if data_type_selection == 'Both':
        filtered_data = combined_data[(combined_data['organisationunitname'] == organization_selection)]
    else:
        filtered_data = combined_data[(combined_data['organisationunitname'] == organization_selection) & (combined_data['Type'] == data_type_selection)]
    
    interactive_selection = st.sidebar.selectbox('Interactive Graph', ['Yes', 'No'])
    
    if st.sidebar.button('Show Trend'):
        st.subheader(f'Trend for {product_selection} - {organization_selection} ({data_type_selection})')
        plot_trends(filtered_data, product_selection)
    
    if st.sidebar.button('Show Forecast'):
        st.subheader(f'Forecast for {product_selection} - {organization_selection} ({data_type_selection})')
        train_and_plot_forecasts(filtered_data, product_selection, interactive_selection)
    
    st.markdown('</div>', unsafe_allow_html=True)

if __name__ == '__main__':
    main()
```

```python
import pandas as pd
import numpy as np
from prophet import Prophet
import streamlit as st
import plotly.graph_objs as go
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Load the datasets
consumption_file_path = 'data (25).csv'
service_file_path = 'data (27).csv'
consumption_data = pd.read_csv(consumption_file_path)
service_data = pd.read_csv(service_file_path)

# Mapping dictionary
mapping = {
    'MOH 747A_Male Condoms': 'Male Condoms',
    'MOH 711 Client receiving Male condoms': 'Male Condoms',
    'MOH 747A_Female Condoms': 'Female Condoms',
    'MOH 711 Clients receiving Female Condoms': 'Female Condoms',
    'MOH 747A_Emergency Contraceptive pills': 'EC Pills',
    'MOH 711 Emergency contraceptive pill': 'EC Pills',
    'MOH 747A_Combined Oral contraceptive Pills': 'COCs',
    'MOH 711 Pills Combined oral contraceptive': 'COCs',
    'MOH 747A_Progestin only pills': 'POPs',
    'MOH 711 Pills progestin only': 'POPs',
    'MOH 747A_Cycle Beads': 'Cycle Beads',
    'MOH 711 Rev 2020_Clients given cycle beads': 'Cycle Beads',
    'MOH 747A_DMPA-IM': 'DMPA-IM',
    'MOH 711 Rev 2020_FP Injections DMPA- IM': 'DMPA-IM',
    'MOH 747A_DMPA-SC': 'DMPA-SC',
    'MOH 711 Rev 2020_FP Injections DMPA- SC': 'DMPA-SC',
    'MOH 747A_Implants (1-Rod) – ENG 68mg': 'Implanon',
    'MOH 711 Rev 2020_Implants insertion 1 Rod': 'Implanon',
    'MOH 711 Rev 2020_Implants insertion 2 Rod': '2 Rod',
    'MOH 747A_Hormonal IUCD': 'Hormonal IUCD',
    'MOH 711 Rev 2020_IUCD Insertion Hormonal': 'Hormonal IUCD',
    'MOH 747A_Non-Hormonal IUCD': 'Non-Hormonal IUCD',
    'MOH 711 Rev 2020_IUCD Insertion Non Hormonal': 'Non-Hormonal IUCD',
    'MOH 747A_Implants (2-Rod) - LNG 75mg (3 years)': 'Levoplant',
    'MOH 747A_Implant (2-Rod) – LNG 75mg (5 years)': 'Jadelle'
}

# Apply the mapping to the columns
consumption_data.rename(columns=mapping, inplace=True)
service_data.rename(columns=mapping, inplace=True)

# Add a column to distinguish between consumption and service
consumption_data['Type'] = 'Consumption'
service_data['Type'] = 'Service'

# Combine the datasets
combined_data = pd.concat([consumption_data, service_data], ignore_index=True)

# Fill missing values
combined_data.fillna(0, inplace=True)

# Convert 'periodname' to datetime and aggregate monthly
combined_data['periodname'] = pd.to_datetime(combined_data['periodname'], format='%b-%y', errors='coerce')

# Aggregate on a monthly level
combined_data = combined_data.groupby([pd.Grouper(key='periodname', freq='M'), 'organisationunitname', 'Type']).sum().reset_index()

# Remove columns that are not needed in the aggregation (non-numeric ones)
non_numeric_columns = combined_data.select_dtypes(exclude=[np.number]).columns.difference(['periodname', 'organisationunitname', 'Type'])
combined_data.drop(non_numeric_columns, axis=1, inplace=True)

# List of unique products to include in the selection
unique_products = [
    'Male Condoms',
    'Female Condoms',
    'EC Pills',
    'COCs',
    'POPs',
    'Cycle Beads',
    'DMPA-IM',
    'DMPA-SC',
    'Implanon',
    '2 Rod',
    'Hormonal IUCD',
    'Non-Hormonal IUCD',
    'Levoplant',
    'Jadelle'
]


# Function to plot trends
def plot_trends(data, product):
    consumption_data = data[data['Type'] == 'Consumption'][['periodname', product]].rename(columns={'periodname': 'ds', product: 'y'})
    service_data = data[data['Type'] == 'Service'][['periodname', product]].rename(columns={'periodname': 'ds', product: 'y'})

    trend_fig = go.Figure()

    # Plot consumption data
    if len(consumption_data) > 0:
        trend_fig.add_trace(go.Scatter(x=consumption_data['ds'], y=consumption_data['y'], mode='lines', name='Consumption Data', line=dict(color='red')))
    else:
        st.warning("No consumption data available.")
    
    # Plot service data
    if len(service_data) > 0:
        trend_fig.add_trace(go.Scatter(x=service_data['ds'], y=service_data['y'], mode='lines', name='Service Data', line=dict(color='blue')))
    else:
        st.warning("No service data available.")
    
    trend_fig.update_layout(title=f'Trend for {product}', xaxis_title='Date', yaxis_title='Quantity')
    st.plotly_chart(trend_fig)

# Streamlit app
# Streamlit app
def main():
    st.set_page_config(layout="wide", page_title="Prophet Forecasting App")
    st.markdown(
        """
        <style>
        .main {
            background-color: #f0f0f5;
            padding: 20px;
            font-family: Arial, sans-serif;
        }
        .forecast-table th {
            font-weight: bold;
            text-align: center;
        }
        .forecast-table td {
            text-align: center;
        }
        </style>
        """,
        unsafe_allow_html=True
    )
    st.markdown('<div class="main">', unsafe_allow_html=True)
    st.title('Prophet Forecasting App')
    
    st.sidebar.header('Configuration')
    organization_selection = st.sidebar.selectbox('Select Organization', combined_data['organisationunitname'].unique())
    product_selection = st.sidebar.selectbox('Select Product', unique_products)
    data_type_selection = st.sidebar.selectbox('Select Data Type', ['Consumption', 'Service', 'Both'])
    
    if data_type_selection == 'Both':
        filtered_data = combined_data[(combined_data['organisationunitname'] == organization_selection)]
    else:
        filtered_data = combined_data[(combined_data['organisationunitname'] == organization_selection) & (combined_data['Type'] == data_type_selection)]
    
    interactive_selection = st.sidebar.selectbox('Interactive Graph', ['Yes', 'No'])
    
    if st.sidebar.button('Show Trend'):
        st.subheader(f'Trend for {product_selection} - {organization_selection} ({data_type_selection})')
        plot_trends(filtered_data, product_selection)
    
    if st.sidebar.button('Show Forecast'):
        st.subheader(f'Forecast for {product_selection} - {organization_selection} ({data_type_selection})')
        train_and_plot_forecasts(filtered_data, product_selection, interactive_selection)
    
    st.markdown('</div>', unsafe_allow_html=True)

if __name__ == '__main__':
    main()

```
